```python
X = data.drop(['heart_disease'],axis=1)
Y = data['heart_disease']
```

```python
Y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X , Y , test_size = 0.2 )#,random_state=5)
```

```python
X_train
```

```python
X_test
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import tree
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
```

```python
lr = LinearRegression()
lr.fit(X_train,Y_train)

lor = LogisticRegression()
lor.fit(X_train,Y_train)

svm = SVR()
svm.fit(X_train,Y_train)

gr = GradientBoostingRegressor()
gr.fit(X_train,Y_train)

km = KMeans(n_clusters=3)
km.fit(X_train,Y_train)
```

```python
Y_pred1 = lr.predict(X_test)

Y_pred2 = lor.predict(X_test)

Y_pred3 = svm.predict(X_test)

Y_pred4 = gr.predict(X_test)

Y_pred5 = km.predict(X_test)
```

```python
df = pd.DataFrame({'Actual':Y_test,'lr':Y_pred1,'lor':Y_pred2,'svm':Y_pred3,'gr':Y_pred4,'km':Y_pred5})
df
```

```python
import matplotlib.pyplot as plt

plt.subplot(331)
plt.plot(df['Actual'].iloc[0:10],label='Actual')
plt.plot(df['lr'].iloc[0:10],label='lr')
plt.legend()

plt.subplot(332)
plt.plot(df['Actual'].iloc[0:10],label='Actual')
plt.plot(df['lor'].iloc[0:10],label='lor')
plt.legend()

plt.subplot(333)
plt.plot(df['Actual'].iloc[0:10],label='Actual')
plt.plot(df['svm'].iloc[0:10],label='svr')
plt.legend()


plt.subplot(334)
plt.plot(df['Actual'].iloc[0:10],label='Actual')
plt.plot(df['gr'].iloc[0:10],label='gr')
plt.legend()

plt.subplot(335)
plt.plot(df['Actual'].iloc[0:10],label='Actual')
plt.plot(df['km'].iloc[0:10],label='km')
plt.legend()
```
